```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(3,4))
ax.hist(Dream.Frequency, bins=3,width=0.5 ,color='orange')
plt.title("Frequency of Dreaming When Sleep")
plt.ylabel("Number of People")
plt.xlabel("Level of Frequencey")
```

```python
fig, ax = plt.subplots(figsize=(3,4))
ax.hist(Dream.Colour, bins=3,width=0.5)
plt.title("Colour Modes of Dreams")
plt.ylabel("Number of People")
plt.xlabel("Colour Mode")
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.hist(Dream.Controlling, bins=3,width=0.5, color='purple')
plt.title("Frequency of able to consciously control behaviour in the dream")
plt.ylabel("Number of People")
plt.xlabel("Level of Frequencey")
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.hist(Dream.World, bins=3,width=0.5, color='blue')
plt.title("Dream based on Real or Surreal world")
plt.ylabel("Number of People")

plt.xlabel("Level of Frequencey")
plt.xticks(rotation=90)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.hist(Dream.Remenber, bins=3,width=0.5, color='green')
plt.title("Frequency of able to remember dream after wake up")
plt.ylabel("Number of People")
plt.xlabel("Level of Frequencey")
plt.xticks(rotation=90)
```

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.hist(Dream.Remenber, bins=3,width=0.5, color='green', alpha=0.5)
ax.hist(Dream.World, bins=3,width=0.5, color='blue',alpha=0.5)
plt.xticks(rotation=90)
```

```python
plt.hist(Dream.Controlling, bins=5,width=0.6, color='red', label='Controlling',alpha=0.6)
plt.hist(Dream.Frequency, bins=4,width=0.5 ,color='yellow',label='DreamFrequency', alpha=0.5)
plt.hist(Dream.Remenber, bins=4,width=0.5, color='blue', label='Remenbering', alpha=0.4)
plt.hist(Dream.Colour, bins=3,width=0.5, color='orange', label='ColourfulDream', alpha=0.5)
plt.hist(Dream.Senses, bins=5,width=0.5, color='green', label='Senses', alpha=0.5)
plt.xticks(rotation=90)
plt.legend()

#commentThe ability to consciously control behaviour in a dream is directly proportional to 
#the sensory experience , the colourfulness, and the memory of the dream.
```
